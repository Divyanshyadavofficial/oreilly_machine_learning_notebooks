```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
X = iris.data[:,2:]
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X,y)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_clf,
    out_file="iris_tree.dot",
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
```

```python
import os
os.system("dot -Tjpeg iris_tree.dot -o iris_tree.jpeg")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("iris_tree.jpeg")
plt.imshow(img)
plt.axis('off')  
plt.show()
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
np.random.seed(42)
X = np.sort(np.random.rand(100,1))
y = X**2 + 0.05 * np.random.randn(100,1)

tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X,y)

X_new = np.linspace(0,1,500).reshape(-1,1)

y_pred = tree_reg.predict(X_new)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(X, y, "b.", label="Training data")
plt.plot(X_new, y_pred, "r-", linewidth=2, label="Prediction")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Decision Tree Regression (max_depth=2)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_reg,
    out_file="reg_tree.dot",
    feature_names=["X"],
    special_characters=True,
    rounded=True,
    filled=True
)
```

```python
import os
os.system("dot -Tjpeg reg_tree.dot -o reg_tree.jpeg")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("reg_tree.jpeg")
plt.imshow(img)
plt.axis('off')  
plt.show()
```
